```python
import pandas as pd
import matplotlib.pyplot as plt

# Replace 'your_file_path.csv' with the actual path of your dataset
file_path = '/content/titanic.csv'

# Load the data
df = pd.read_csv(file_path)

# Remove rows where 'Survived' is 0
survived_df = df[df['Survived'] == 1]

# Count the number of male and female survivors
survived_counts = survived_df['Sex'].value_counts()

# Plot the pie chart with counts and legend
plt.figure(figsize=(8, 8))
plt.pie(survived_counts, labels=survived_counts.index, autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightcoral'])
plt.title('Gender Ratio Among Survivors')

# Add legend with counts
legend_labels = [f"{sex} ({count})" for sex, count in survived_counts.items()]
plt.legend(legend_labels, loc='upper left', bbox_to_anchor=(0.05, 0.95))

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Replace 'your_file_path.csv' with the actual path of your dataset
file_path = '/content/titanic.csv'

# Load the data
df = pd.read_csv(file_path)

# Create age groups
bins = [0, 18, 35, 50, float('inf')]
labels = ['<18', '18-35', '35-50', '>50']

# Apply age groups to the dataset
df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# Plot the bar chart with different colors
colors = {'<18': 'green', '18-35': 'orange', '35-50': 'blue', '>50': 'red'}
fig, ax = plt.subplots()

for age_group, color in colors.items():
    age_group_data = df[(df['Age_Group'] == age_group) & (df['Survived'] == 1)]
    count = age_group_data['Age_Group'].count()

    if count > 0:
        bars = ax.bar(age_group, count, color=color, label=f'{age_group} ')

# Customize the plot
ax.set_title('Survival Based on Age')
ax.set_xlabel('Age')
ax.set_ylabel('Count')
ax.legend()

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Paste the full path of your dataset file here
file_path = '/content/titanic.csv'

# Load the dataset into a Pandas DataFrame
df = pd.read_csv(file_path)

# Exclude rows with 'Survived' equal to 1
df_filtered = df[df['Survived'] != 1]

# Create a heatmap for 'Survived' and 'Pclass'
heatmap_data = pd.crosstab(df_filtered['Survived'], df_filtered['Pclass'], normalize='index')

# Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(heatmap_data, annot=True, cmap='coolwarm', fmt='.2f', cbar_kws={'label': 'Proportion'})

# Add legend to the x-axis
class_labels = {1: 'Upper Class', 2: 'Middle Class', 3: 'Lower Class'}
plt.xticks(ticks=[0.5, 1.5, 2.5], labels=[class_labels[label] for label in heatmap_data.columns], rotation=0)

plt.title('Death ratio of Titanic based on Passenger class')
plt.xlabel('Passenger Class')
plt.ylabel('Dead')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Paste the full path of your dataset file here
file_path = '/content/titanic.csv'

# Load the dataset into a Pandas DataFrame
df = pd.read_csv(file_path)


# Create a scatter plot
plt.figure(figsize=(10, 6))
scatter = sns.scatterplot(data=df, x='Pclass', y='Fare', hue='Pclass', palette={1: 'green', 2: 'red', 3: 'orange'})

# Add hover information using text annotations
for line in range(0, df.shape[0]):
    scatter.text(df['Pclass'][line] + 0.05, df['Fare'][line], f'Fare: {df["Fare"][line]:.2f}', horizontalalignment='left', size='small', color='black')

# Add a legend box
plt.legend(title='Passenger Class', loc='upper right', bbox_to_anchor=(1.2, 1))

plt.title('Scatter Plot of Fare by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()
```
